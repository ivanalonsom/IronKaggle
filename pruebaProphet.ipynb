```python
import pandas as pd

from prophet import Prophet

df_ini = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/sales.csv")

X_test = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/ironkaggle_notarget.csv")

df_real_data = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/ironkaggle_solutions.csv")
```

### PRUEBA DE PROPHET

```python
df_ini.set_index('True_index', inplace=True)

df_prophet = df_ini[["Date", "Sales"]]
```

```python
df_prophet.rename(columns={'Date': 'ds', 'Sales': 'y'}, inplace=True)
```

```python
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'], format='%Y-%m-%d')
```

```python
m = Prophet()
m.fit(df_prophet)
```

```python
X_test_pred = X_test[["Date"]]
X_test_pred.rename(columns={'Date': 'ds'}, inplace=True)
X_test_pred['ds'] = pd.to_datetime(X_test_pred['ds'], format='%Y-%m-%d')
```

```python
pred = m.predict(X_test_pred)
```

```python
pred
```

```python
X_test_pred["yhat"] = pred["yhat"]
```

```python
X_test_pred
```

```python
df_real_data.merge(X_test_pred["yhat"], how="left", left_index=True, right_index=True)
```
